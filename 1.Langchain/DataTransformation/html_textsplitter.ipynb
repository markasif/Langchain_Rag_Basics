```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_strings="""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>About AI</title>
    <style>
        body { font-family: Arial, sans-serif; text-align: center; margin: 50px; }
        h1 { color: #007bff; }
        p { font-size: 18px; }
    </style>
</head>
<body>
    <h1>Artificial Intelligence</h1>
    <p>Artificial Intelligence (AI) is the simulation of human intelligence in machines that are programmed to think and learn.</p>
    <p>AI is used in various fields such as healthcare, finance, robotics, and more.</p>
    <br>
    <p>ai is changing the world"</p>
    <p> the world is going to be control of automation"</p>
    <h2>Machine leaninh</h2>
    <p>Machine Learning (ML)</strong> is a branch of AI where computers learn from data and improve performance without explicit programming.</p>
    <h3>deep learning</h3>
    <p><strong>Deep Learning (DL)</strong> is a subset of ML that uses neural networks with multiple layers to process complex patterns in large datasets.</p>
</body>
</html>
"""

header_to_split =[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")

]
html_splitter=HTMLHeaderTextSplitter(header_to_split)
html_header_splits=html_splitter.split_text(html_strings)
html_header_splits
```

```python
url="https://en.wikipedia.org/wiki/Kerala"

header =[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3"),
    ("h4","Header 4")
]

html_split=HTMLHeaderTextSplitter(header)
splits = html_split.split_text_from_url(url)
splits
```
